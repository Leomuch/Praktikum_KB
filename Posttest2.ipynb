Nomor 1

Rata-rata, standar deviasi, dan kuantil (Q1, Q2, Q3) dari atribut numerik memberikan gambaran umum tentang distribusi data, seperti sebaran harga atau rating produk.

```python
# Memilih hanya kolom numerik
df_Muchlas = Muchlas.select_dtypes(include=['number'])

# Menghitung statistik deskriptif
statistik = df_Muchlas.describe()

# Menghitung statistik deskriptif
statistik = df_Muchlas.describe()

# Mengambil statistik yang diperlukan
rata_rata = statistik.loc['mean']
std_deviasi = statistik.loc['std']
nilai_minimum = statistik.loc['min']
q1 = df_Muchlas.quantile(0.25)
median = statistik.loc['50%']
q3 = df_Muchlas.quantile(0.75)
nilai_maksimum = statistik.loc['max']

# Membuat DataFrame untuk hasil
tabel_statistik = pd.DataFrame({
    'Rata-rata': rata_rata,
    'Standar Deviasi': std_deviasi,
    'Nilai Minimum': nilai_minimum,
    'Q1': q1,
    'Median (Q2)': median,
    'Q3': q3,
    'Nilai Maksimum': nilai_maksimum
})

# Menampilkan tabel
print(tabel_statistik)
```

Nomor 3

Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
data = {}

data['Jumlah Record'] = Muchlas.shape[0]
data['Jumlah Attribute'] = Muchlas.shape[1]
data['Nama Attribute'] = Muchlas.columns.tolist()
data['Jumlah Attribute Dengan Tipe Angka'] = Muchlas.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(data.keys())

table = pretty.PrettyTable()
table.add_column(dictKeys[0], [data['Jumlah Record']])
table.add_column(dictKeys[1], [data['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(dictKeys[3], [data['Jumlah Attribute Dengan Tipe Angka']])
print(table)
```

Nomor 5

Visualisasikan dataset dengan Histogram

```python
# Plot Histogram
# Visualize the distribution of 'Age'
plt.figure(figsize=(8, 6))
plt.hist(Muchlas['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Football Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Visualisasikan dataset Bar Chart

```python
# Plot Bar Chart
position_counts = Muchlas['position'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(position_counts.index, position_counts.values, color='skyblue')
plt.title('Count of Football Players Based on Position')
plt.xlabel('Position')
plt.ylabel('Count')

# Memiringkan label x agar mudah dibaca
plt.xticks(rotation=80)
plt.show()
```

Tampilkan korelasi dataset menggunakan heatmap.

```python
# Menghitung korelasi
korelasi = df_Muchlas.corr()

# Membuat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Heatmap Korelasi Dataset')
plt.show()
```
